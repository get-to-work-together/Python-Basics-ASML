# Demo Pandas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename = 'Sandbox/ca-500.csv'
```

```python
df = pd.read_csv(filename, sep=';')
df
```

```python
df.info()
```

```python
counts = df['province'].value_counts()
counts
```

```python
plt.barh(counts.index, counts.values)
plt.title('Provinces')
plt.show()
```

```python
df_selected = df[['first_name','last_name','city','email']].query('city=="Montreal"')
df_selected
```

```python
df_selected.info()
```

```python
df_selected.sort_values('last_name')
```

```python
for s in df_selected.itertuples():
    print(s.first_name, s.city)
```

```python
df_selected.apply(lambda r: r.first_name + ' ' + r['last_name'])
```
